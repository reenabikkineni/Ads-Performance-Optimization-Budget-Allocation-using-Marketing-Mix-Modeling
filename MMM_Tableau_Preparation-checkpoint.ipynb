
# Marketing Mix Modeling (MMM) – Tableau Preparation Notebook

This notebook prepares an **analytics-ready dataset** for Tableau dashboards.

Workflow:
Raw MMM Dataset → Feature Engineering (Python) → KPI Creation → Tableau Dashboard Dataset

The goal is to create clean fields so Tableau requires **minimal calculations**.


## Step 1 — Load Dataset

```

import pandas as pd

# Load dataset (make sure CSV is in the same folder)
df = pd.read_csv("Multi‑Region Ecommerce MMM Dataset.csv")

df.head()
```

## Step 2 — Convert Date Column

```

# Convert date column (update column name if needed)
df['DATE'] = pd.to_datetime(df['DATE'])

df.info()
```

## Step 3 — Create Total Marketing Spend

```

df['total_spend'] = (
    df['GOOGLE_PAID_SEARCH_SPEND'] +
    df['META_FACEBOOK_SPEND'] +
    df['META_INSTAGRAM_SPEND'] +
    df['TIKTOK_SPEND'] +
    df['GOOGLE_DISPLAY_SPEND']
)

df[['total_spend']].head()
```

## Step 4 — Create Channel Groups (Optional but Helpful)

```

df['paid_social_spend'] = df['META_FACEBOOK_SPEND'] + df['META_INSTAGRAM_SPEND']
df['google_spend'] = df['GOOGLE_PAID_SEARCH_SPEND'] + df['GOOGLE_DISPLAY_SPEND']

df[['paid_social_spend','google_spend']].head()
```

## Step 5 — Create Conversion Metrics

```

# Update clicks/sessions column name if different
df['conversion_rate'] = df['ALL_PURCHASES'] / df['SESSIONS']

df[['conversion_rate']].describe()
```

## Step 6 — Marketing Efficiency Metrics (For Tableau KPIs)

```

df['cost_per_purchase'] = df['total_spend'] / df['ALL_PURCHASES']
df['purchase_per_dollar'] = df['ALL_PURCHASES'] / df['total_spend']

df[['cost_per_purchase','purchase_per_dollar']].head()
```

## Step 7 — Create Time-Based Fields (Very Useful in Tableau)

```

df['year'] = df['DATE'].dt.year
df['month'] = df['DATE'].dt.month
df['year_month'] = df['DATE'].dt.to_period('M').astype(str)

df[['year','month','year_month']].head()
```

## Step 8 — Select Tableau Analytics Columns

```

tableau_df = df[[
    'DATE',
    'year',
    'month',
    'year_month',
    'TERRITORY',
    'total_spend',
    'ALL_PURCHASES',
    'conversion_rate',
    'cost_per_purchase',
    'purchase_per_dollar',
    'GOOGLE_PAID_SEARCH_SPEND',
    'GOOGLE_DISPLAY_SPEND',
    'META_FACEBOOK_SPEND',
    'META_INSTAGRAM_SPEND',
    'TIKTOK_SPEND'
]]

tableau_df.head()
```

## Step 9 — Export Tableau Dataset

```

tableau_df.to_csv("mmm_tableau_dataset.csv", index=False)

print("✅ Tableau dataset exported successfully!")
```
