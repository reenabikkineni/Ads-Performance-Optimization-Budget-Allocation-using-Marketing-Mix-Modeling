
# Marketing Mix Modeling (MMM) – Tableau Preparation Notebook

This notebook prepares an **analytics-ready dataset** for Tableau dashboards using the
**Multi‑Region Ecommerce MMM Dataset**.

Workflow:
Raw Marketing Data → Feature Engineering (Python) → KPI Creation → Tableau Dashboard Dataset

The goal is to create clean analytical fields so Tableau dashboards require minimal calculations.


## Step 1 — Import Libraries & Load Dataset

```

import pandas as pd

# Load dataset
df = pd.read_csv("Multi-Region Ecommerce MMM Dataset.csv")

df.head()
```

## Step 2 — Convert Date Column

```

df['DATE_DAY'] = pd.to_datetime(df['DATE_DAY'])

df.info()
```

## Step 3 — Create Total Marketing Spend

```

df['total_spend'] = (
    df['GOOGLE_PAID_SEARCH_SPEND'] +
    df['GOOGLE_SHOPPING_SPEND'] +
    df['GOOGLE_PMAX_SPEND'] +
    df['GOOGLE_DISPLAY_SPEND'] +
    df['GOOGLE_VIDEO_SPEND'] +
    df['META_FACEBOOK_SPEND'] +
    df['META_INSTAGRAM_SPEND'] +
    df['META_OTHER_SPEND'] +
    df['TIKTOK_SPEND']
)

df[['total_spend']].head()
```

## Step 4 — Channel Grouping

```

# Google ecosystem spend
df['google_total_spend'] = (
    df['GOOGLE_PAID_SEARCH_SPEND'] +
    df['GOOGLE_SHOPPING_SPEND'] +
    df['GOOGLE_PMAX_SPEND'] +
    df['GOOGLE_DISPLAY_SPEND'] +
    df['GOOGLE_VIDEO_SPEND']
)

# Meta ecosystem spend
df['meta_total_spend'] = (
    df['META_FACEBOOK_SPEND'] +
    df['META_INSTAGRAM_SPEND'] +
    df['META_OTHER_SPEND']
)

df[['google_total_spend','meta_total_spend']].head()
```

## Step 5 — Create Paid Click Metrics

```

df['total_paid_clicks'] = (
    df['GOOGLE_PAID_SEARCH_CLICKS'] +
    df['GOOGLE_SHOPPING_CLICKS'] +
    df['GOOGLE_PMAX_CLICKS'] +
    df['GOOGLE_DISPLAY_CLICKS'] +
    df['GOOGLE_VIDEO_CLICKS'] +
    df['META_FACEBOOK_CLICKS'] +
    df['META_INSTAGRAM_CLICKS'] +
    df['META_OTHER_CLICKS'] +
    df['TIKTOK_CLICKS']
)

df[['total_paid_clicks']].head()
```

## Step 6 — Organic & Owned Traffic Metrics

```

df['organic_total_clicks'] = (
    df['DIRECT_CLICKS'] +
    df['BRANDED_SEARCH_CLICKS'] +
    df['ORGANIC_SEARCH_CLICKS'] +
    df['EMAIL_CLICKS'] +
    df['REFERRAL_CLICKS'] +
    df['ALL_OTHER_CLICKS']
)

df[['organic_total_clicks']].head()
```

## Step 7 — Conversion Metrics

```

df['conversion_rate'] = df['ALL_PURCHASES'] / df['total_paid_clicks']

df[['conversion_rate']].describe()
```

## Step 8 — Marketing Efficiency Metrics

```

df['cost_per_purchase'] = df['total_spend'] / df['ALL_PURCHASES']
df['purchase_per_dollar'] = df['ALL_PURCHASES'] / df['total_spend']

df[['cost_per_purchase','purchase_per_dollar']].head()
```

## Step 9 — Time Intelligence Fields

```

df['year'] = df['DATE_DAY'].dt.year
df['month'] = df['DATE_DAY'].dt.month
df['year_month'] = df['DATE_DAY'].dt.to_period('M').astype(str)

df[['year','month','year_month']].head()
```

## Step 10 — Prepare Tableau Dataset

```

tableau_df = df[[
    'DATE_DAY',
    'year',
    'month',
    'year_month',
    'TERRITORY_NAME',
    'ORGANISATION_VERTICAL',
    'total_spend',
    'google_total_spend',
    'meta_total_spend',
    'TIKTOK_SPEND',
    'ALL_PURCHASES',
    'FIRST_PURCHASES',
    'total_paid_clicks',
    'organic_total_clicks',
    'conversion_rate',
    'cost_per_purchase',
    'purchase_per_dollar'
]]

tableau_df.head()
```

## Step 11 — Export Tableau Dataset

```

tableau_df.to_csv("mmm_tableau_dataset.csv", index=False)

print("✅ Tableau dataset ready!")
```
